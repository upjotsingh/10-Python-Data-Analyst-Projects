```python

# 3. Counting Transactions per Customer

### Step 1: Check one customer's transactions  
Print all transactions for a single customer (e.g., Alice).

### Step 2: Create a dictionary for counts  
Make an empty dictionary to store how many transactions each customer has.

### Step 3: Loop and count  
Go through each transaction in the list.  
For every transaction, check the `customer` name:
- If the customer already exists in the dictionary, increase their count by 1.
- If not, add them to the dictionary with a starting count of 1.

### Step 4: Print results  
Show each customer's name and their total number of transactions.


````python
# Step 1: Look at one customer's transactions first
frankTrnsx=[]
for item in transactions:
    if(item['customer']=='Frank'):
        frankTrnsx.append(item)


# Step 2: Create an empty dictionary for counts
cxDict={}


# Step 3: Count transactions for each customer
for uCx in uniqueCustomers:
    cxDict[uCx]=0
    for tx in transactions:
        if(tx['customer']==uCx):
            cxDict[uCx]+=1


# Step 4: Print results
print('Frank Transactions: ',frankTrnsx)
print('\nNo. of Transactions for each customer: ',cxDict)
````

# 4. Find Duplicate Transactions

Check the dataset for exact duplicate transactions (same customer, type, amount, and branch).

````python
# Removing duplicate transactions
seen=set()
duplicateTxs=[]
for tx in transactions:
    if tuple(tx.items()) in seen:
        duplicateTxs.append(tx)
    else:
        seen.add(tuple(tx.items()))
print('Duplicate Transactions:',duplicateTxs)
````

# 5. Clean and Summarize Customer Activity

Compute the **net balance** (deposits minus withdrawals) for each customer and print the results in a neat table.

````python
# Step 1: Build valid_transactions

filteredTransx=[]
for item in transactions:
    if(isinstance(item['amount'],(int,float))):
        filteredTransx.append(item)

seen=set()
duplicateTxs=set()
for tx in filteredTransx:
    if tuple(tx.items()) in seen:
        duplicateTxs.add(tuple(tx.items()))
    else:
        seen.add(tuple(tx.items()))
        
valid_transactions=[]
for tx in filteredTransx:
    if tuple(tx.items()) not in duplicateTxs:
        valid_transactions.append(tx)

newList=[]
for cx in uniqueCustomers:
    totalDeposit=0
    totalWithdrawal=0
    for tx in valid_transactions:
        if(cx==tx['customer']):
            if(tx['type']=='withdrawal'):
                totalWithdrawal+=tx['amount']
            elif(tx['type']=='deposit'):
                totalDeposit+=tx['amount']
    netBalance=totalDeposit-totalWithdrawal
    newList.append({'customer':cx,'netBalance':netBalance,'totalDeposit':totalDeposit,'totalWithdrawal':totalWithdrawal})


# Step 2: Build summary
print(f"{'Name':<10} | {'Total Deposits':>5} | {'Total Withdrawal':>5} | {'Net Balance':>5}")
for item in newList:
    print(f"{item['customer']:<10} | {item['totalDeposit']:>14} | {item['totalWithdrawal']:>16} | {item['netBalance']:>10}")

print('\nDavid has negative net balance, could be considered as credit or wrong data.')

````
