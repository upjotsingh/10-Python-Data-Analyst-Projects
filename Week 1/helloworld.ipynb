## Let's visualize a bit

#### Plotting two time series in a 2 × 1 grid of subplots

```python
# Example data
time = [1, 2, 3, 4, 5]
series1 = [10, 15, 20, 18, 22]
series2 = [5, 7, 9, 6, 8]

# Create a 2 × 1 grid of subplots
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# Top subplot
axes[0].plot(time, series1, marker='o')
axes[0].set_title("Series 1 Over Time")
axes[0].set_ylabel("Value")

# Bottom subplot
axes[1].plot(time, series2, marker='s', color='green')
axes[1].set_title("Series 2 Over Time")
axes[1].set_xlabel("Time")
axes[1].set_ylabel("Value")

plt.tight_layout()  # Adjust spacing to avoid overlap
plt.show()
```

```python
 # Creating the Tips dataset
np.random.seed(42)
n_samples = 200
days = ['Thur', 'Fri', 'Sat', 'Sun']
times = ['Lunch', 'Dinner']
genders = ['Male', 'Female']

# Generating random data for the Tips dataset
tips = pd.DataFrame({
    'total_bill': np.round(np.random.uniform(5, 50, n_samples), 2),
    'tip': np.round(np.random.uniform(1, 10, n_samples), 2),
    'sex': np.random.choice(genders, n_samples),
    'smoker': np.random.choice(['Yes', 'No'], n_samples),
    'day': np.random.choice(days, n_samples),
    'time': np.random.choice(times, n_samples),
    'size': np.random.randint(1, 6, n_samples)
})
```

```python
# Visualizing the relationship between total bill and tip using a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
import matplotlib.pyplot as plt
# Data
x = [1, 2, 3, 4]  # X-axis data
y = [10, 20, 25, 30]  # Y-axis data
# Creating the plot
plt.figure(figsize=(4, 2))  # Creates the figure with specific size
plt.plot(x, y)  # Plots the data on the figure
# Adding components
plt.title("Basic Line Plot")  # Adding a title
plt.xlabel("X-axis")  # Labeling the X-axis
plt.ylabel("Y-axis")  # Labeling the Y-axis
# Display the plot
plt.show()  # Displaying the plot
```

```python
x = [1, 2, 3, 4]  # X-axis data
y = ['a', 'e', 'b', 'c']  # Y-axis data
plt.plot(x, y, color='green', linestyle='--', marker=[]'o')  # Creating a line plot with green color, dashed line, and circle markers
plt.title("Line Plot Example")  # Adding a title
plt.xlabel("X-axis")  # Labeling the X-axis
plt.ylabel("Y-axis")  # Labeling the Y-axis
plt.show()  # Displaying the plot
```
